```python
import tensorflow as tf
import numpy as np
from tfRegression import TfRegression
```

```python
n_sample = 1000
learning_rate=0.01
num_epochs=100
random_X = np.random.normal(size=(n_sample, 4))
random_Y = np.random.normal(size=(n_sample, 1))
w =tf.Variable(tf.random_normal((5, 1), name="weight", dtype=tf.float64))
```

```python
regression = TfRegression(inputs=random_X, output= random_Y)
```

```python
init = tf.global_variables_initializer()
x_ = tf.placeholder(tf.float32, shape=[None, 4 + 1])
y_ = tf.placeholder(tf.float32, shape=[None, 1])


inputs_ = regression.append_bias(random_X)
w = tf.Variable(tf.random_normal((4 + 1, 1),
                                         name="weight", dtype=tf.float64))


y_pred= regression.model(inputs_, w)
mse = tf.reduce_mean(tf.square(random_Y - y_pred))
optimizer = tf.train.GradientDescentOptimizer(
    learning_rate)
train_op = optimizer.minimize(mse)

cost_trace = []
with tf.Session() as sess:
    # Setup a session & initialize all variables
    sess.run(tf.global_variables_initializer())
    for i in range(num_epochs):
        for x_batch, y_batch in next_batch(random_X, random_Y, 100):
            x_batch_plus_intercept = regression.append_bias(x_batch)
            
            cost_trace.append(sess.run(mse, feed_dict={x_: x_batch_plus_intercept, y_: y_batch}))
            sess.run(train_op, feed_dict={x_: x_batch_plus_intercept, y_: y_batch})
    print(w.eval())
```
