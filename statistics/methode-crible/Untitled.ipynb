```python
from point import Point
from circle import Circle
from rectangle import Rectangle


import matplotlib.pyplot as plt
import matplotlib.patches as patches

%matplotlib inline
```

```python
p1 = Point(2., 2.)
p2 = Point(-2., 3.)
c1 = Circle(p1, 3.)
c2 = Circle(p2, 6.)

r = Rectangle(xmin=-8, ymin=-3, xmax=8, ymax=3)

N = 1000
```

```python
fig, ax = plt.subplots(1, figsize=(30,30))

c1_plt = plt.Circle((p1.x, p1.y), 3, color ='r',alpha=0.2)
c2_plt = plt.Circle((p2.x, p2.y), 6, color = 'g', alpha=0.3)

ax.set_xlim((-30,30))
ax.set_ylim((-30,30))
ax.patch.set_alpha(0.1)

# Draw Rectangle

post_min, post_max = r.framing_rectangle(c1, c2)
rect_frame = [ j-i for i, j in zip(post_min,post_max)]

rect_config = dict(xy=post_min, width = rect_frame[0], height=rect_frame[1])
rect = patches.Rectangle(linewidth=1,edgecolor='r',facecolor='none',**rect_config)

ax.add_artist(c1_plt)
ax.add_artist(c2_plt)
ax.add_patch(rect)

cpt = 0


for i in range(N + 1):
    p = r.random_point()
    # print(f"Does Point: {p} belongs to Circle :{c1} & Circle :{c2}")
    #print((c1.belongs(p) and c2.belongs(p)))
    
    scat = ax.scatter(p.x, p.y)
    fig.canvas.draw_idle()
    if (c1.belongs(p) & c2.belongs(p)):
        cpt += 1


print(f"Intersection ares estimation {(cpt / N) * r.surface()}")
```
